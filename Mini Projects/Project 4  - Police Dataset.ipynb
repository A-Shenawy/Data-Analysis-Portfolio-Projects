# Data Cleaning

```python
df.describe(include = 'all')
df.shape
df.isnull().sum()
```

```python
df.drop('country_name', axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.info()
```

```python
for col in df.columns:
    if df[col].dtypes == "float" or df[col].dtypes == "int" :
        df[col].replace(np.nan, df[col].mean(), inplace = True)
    elif df[col].dtypes == "object":
        df[col].replace(np.nan, df[col].value_counts().idxmax(), inplace = True)
    elif df[col].dtypes == "bool":
        df[col].replace(np.nan, df[col].value_counts().idxmax(), inplace = True)
    else:
        df[col] = df[col]
    
```

```python
## another sol to fill nulls using fillna

for col in df.columns:
    if df[col].dtypes == "float" or df[col].dtypes == "int" :
        df[col].fillna(df[col].mean(), inplace = True)
    elif df[col].dtypes == "object":
        df[col].fillna(df[col].value_counts().idxmax(), inplace = True)
    elif df[col].dtypes == "bool":
        df[col].fillna(df[col].value_counts().idxmax(), inplace = True)
    else:
        df[col] = df[col]
```
